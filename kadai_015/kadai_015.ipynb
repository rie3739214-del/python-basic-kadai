```python
# Humanクラスを定義
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # name と age を標準出力するメソッド
    def printinfo(self):
        print(f"名前：{self.name}")
        print(f"年齢：{self.age}")

# Humanクラスのインスタンスを変数に代入
human1 = Human("佐上 理恵", 35)

# メソッドを呼び出して表示
human1.printinfo()
```
