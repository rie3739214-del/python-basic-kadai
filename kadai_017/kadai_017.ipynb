```python
# Humanクラスを定義
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # 大人かどうかを判定して標準出力するメソッド
    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name} は大人です。")
        else:
            print(f"{self.name} は大人ではありません。")

# Humanクラスのインスタンスを複数生成
human_list = [
    Human("太郎", 25),
    Human("花子", 18),
    Human("次郎", 20)
]

# リストの要素数分だけ check_adult メソッドを呼び出す
for human in human_list:
    human.check_adult()
```
